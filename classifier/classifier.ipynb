```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.metrics import accuracy_score
```

```python
sample_file_path="./samples.csv"
df_site=pd.read_csv(sample_file_path)
```

```python
df_site
```

```python
from sklearn.model_selection import train_test_split
```

```python
X, y = df_site.iloc[:, 3:-2].values, df_site.iloc[:,-1].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0, stratify=y)
```

```python
X_train, X_test, y_train, y_test
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mms=MinMaxScaler()
```

```python
X_train_norm=mms.fit_transform(X_train) # 最大-最小特征缩放
```

```python
X_test_norm=mms.fit_transform(X_test)
```

```python
X_test_norm
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
stdsc=StandardScaler()
X_train_std=stdsc.fit_transform(X_train)
X_test_std=stdsc.fit_transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression(C=1.0,random_state=1,solver='lbfgs',multi_class='multinomial')
lr.fit(X_train_std, y_train)
y_pred=lr.predict(X_test_std)
print('accuracy_score: %.3f' % accuracy_score(y_test, y_pred))
```

```python
from sklearn.svm import SVC
```

```python
svm = SVC(kernel='rbf', random_state=1, gamma='scale', C=1.0)
svm.fit(X_train_std, y_train)
y_pred = svm.predict(X_test_std)
print('accuracy_score: %.3f' % accuracy_score(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree_model=DecisionTreeClassifier(criterion='gini',max_depth=5,random_state=100)
tree_model.fit(X_train_std,y_train)
y_pred = tree_model.predict(X_test_std)
print('accuracy_score: %.3f' % accuracy_score(y_test, y_pred))
```
